# Dictionary to dataFrame


```python
weather_data  = {
    'date' :['1/12/2017','23/6/2018', '15/9/2017'],
    'temperature' :[23,32,36],
    'windspeed' :[6,8,5],
    'event' :['snow ', 'rain','sunny']
}
df  = pd.DataFrame(weather_data)
df
```

```python
df.columns
```

```python
new_df = df[['PDT', 'Max TemperatureF', 'Min TemperatureF',
        'Max Humidity',' Max Wind SpeedMPH', ' CloudCover', ' Events',  'ZIP']]
```

```python
new_df = new_df.rename(columns={'PDT':'date', 'Max TemperatureF':'maxtemp', 'Min TemperatureF':'mintemp', 'Max Humidity':'humidity',
       ' Max Wind SpeedMPH':'windspeed', ' CloudCover':'coudcover', ' Events':'events', 'ZIP':'pin'})
```
