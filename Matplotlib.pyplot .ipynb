```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
%matplotlib inline
plt.plot(x,y)
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(x,y)
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(x,y,'d--')
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(x,y,'+--')
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(x,y, linestyle = '-' ,color='red',marker='d' , markersize=10 )
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(x,y, linestyle = '--' ,color='red',marker='d' , markersize=10 )
```

```python

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Tempareture')
plt.title("Weather")
plt.plot(days,max_t ,label='max')
plt.plot(days,min_t ,label='min')
plt.plot(days,avg_t ,label='avg')
```

```python
plt.xlabel('Days')
plt.ylabel('Tempareture')
plt.title("Weather")
plt.plot(days,max_t ,label='max')
plt.plot(days,min_t ,label='min')
plt.plot(days,avg_t ,label='avg')
plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Tempareture')
plt.title("Weather")
plt.plot(days,max_t ,label='max')
plt.plot(days,min_t ,label='min')
plt.plot(days,avg_t ,label='avg')
plt.legend(loc='upper right' , fontsize='large')
```
