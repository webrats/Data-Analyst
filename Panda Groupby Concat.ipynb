# groupby() 
https://pandas.pydata.org/docs/reference/groupby.html?highlight=groupby%20object%20method

GroupBy.__iter__()

Groupby iterator.

GroupBy.groups

Dict {group name -> group labels}.

GroupBy.indices

Dict {group name -> group indices}.

GroupBy.get_group(name[, obj])

Construct DataFrame from group with provided name.

```python
g = df.groupby('events')
g
```

```python
g.get_group('Fog').head()
```

```python
for key,value in g:
    print(key)
    print(value.head())
```

```python
g.max()
```

```python
g.min()
```

```python
g.describe()
```

```python
for key,value in g:
    print(key)
    print(value.describe())
```
